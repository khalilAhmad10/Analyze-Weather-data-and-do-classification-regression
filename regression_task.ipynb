In the below section data is prepared for regression task

```python
# data is saved in n_red variable
n_reg = prep_data
```

```python
# non_integer features are converted to integers
n_reg['WindGustDir']=pd.factorize(n.WindGustDir)[0]
n_reg['WindDir9am']=pd.factorize(n.WindDir9am)[0]
n_reg['WindDir3pm']=pd.factorize(n.WindDir3pm)[0]
n_reg['RainToday']=pd.factorize(n.RainToday)[0]
n_reg['RainTomorrow']=pd.factorize(n.RainTomorrow)[0]
```

```python
# here the output column is removed from the features data for classification problem
x_reg = n_reg.drop(labels='Evaporation', axis=1)
y_reg = n.iloc[:,3]
```

```python
x_reg.head()
```

```python
y_reg.head()
```

```python
# the data is divided into train and test data
RX_train, RX_test, Ry_train, Ry_test = train_test_split(x_reg, y_reg, test_size=0.2, random_state=42)
```

```python
# in this cell the training data is scalled using min-max scaling method

RX_train = RX_train - RX_train.min()
RX_train /= (RX_train.max() - RX_train.min())
RX_train
```

```python
# in this cell the testing data is scalled using min-max scaling method


RX_test =  RX_test -  RX_test.min()
RX_test /= ( RX_test.max() -  RX_test.min())
RX_test
```

```python
# import important libraries for scores 
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
from sklearn import linear_model, metrics
```

In the below section Linear regression is implemented for finding the regression output

```python
# save the model in a variable and fit it
reg = linear_model.LinearRegression()
reg.fit(RX_train, Ry_train)
```

```python
# predict output through linear regression
lin_reg_pred = reg.predict(RX_test)
```

```python
# mse, rmse and mae are printed for linear regression
print('mean square error is', mean_squared_error(lin_reg_pred,Ry_test))
print('root mean square error is', np.sqrt(mean_squared_error(lin_reg_pred,Ry_test)))
print('mean absolute error is', mean_absolute_error(lin_reg_pred,Ry_test))
```

# In this section polynomial regression is implemented for regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
from tensorflow import keras
from sklearn import preprocessing

# the data is converted to the required input format for polynomial regression

X_train_n =  preprocessing.normalize(RX_train , axis=1)
X_test_n  =  preprocessing.normalize(RX_test , axis=1)
y_train_n = Ry_train

lin_reg_p = LinearRegression()
poly_features = PolynomialFeatures(degree=5, include_bias=False)
X_train_poly = poly_features.fit_transform(X_train_n)
X_test_poly = poly_features.transform(X_test_n)

# fit the model
lin_reg_p.fit(X_train_poly, y_train_n)


```

```python
# predict the output through polynomial regression
poly_reg_pred = lin_reg_p.predict(X_test_poly)
```

```python
# print results for polynomial regression
print('mean square error is', mean_squared_error(poly_reg_pred,Ry_test))
print('root mean square error is', np.sqrt(mean_squared_error(poly_reg_pred,Ry_test)))
print('mean absolute error is', mean_absolute_error(poly_reg_pred,Ry_test))
```
